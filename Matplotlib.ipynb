### Adicinando Linha Horizontal no Gráfico

```python
randomico = np.random.randn(100)
```

### Plotando Gráficos na Mesma Figura

#### Gráfico Horizontal

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 5), sharey = True)
cu = ax0.scatter(range(len(randomico)), randomico);
ax0.axhline(randomico.mean());
ax0.axhline(randomico.std(), color = 'r');
ax0.set(title = "Distribuição Normal Qualquer", ylabel = "Normal_dist")
ax0.legend(*cu.legend_elements(), title = "Teste"); #*cu.legend_elements() não vai funcionar neste plot.

ax1.scatter(range(len(randomico)), np.sin(randomico)); #Só uma brincadeira.
ax1.axhline(np.sin(randomico).mean());
ax1.axhline(np.sin(randomico).std(), color = 'r');

fig.suptitle("Testando", fontsize = 16, fontweight = 'bold');
```

#### Gráfico Vertical

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10), sharex = True)
cu = ax0.scatter(range(len(randomico)), randomico);
ax0.axhline(randomico.mean(), linestyle = ':');
ax0.axhline(randomico.std(), color = 'r', linestyle = '-');
ax0.set(title = "Distribuição Normal Qualquer", ylabel = "Normal_dist")
ax0.legend(*cu.legend_elements(), title = "Teste"); #*cu.legend_elements() não vai funcionar neste plot.
ax0.legend(['mean', 'standard_var', 'dist'])

ax1.scatter(range(len(randomico)), np.sin(randomico), cmap = 'plasma'); #Não dá pra usar cmap sem ter setado c antes.
ax1.axhline(np.sin(randomico).mean(), linestyle = '-.');
ax1.axhline(np.sin(randomico).std(), color = 'r', linestyle = '--');
ax1.set(xlabel = 'x', ylabel = 'sin(dist)')
ax1.legend(['mean', 'standard_var', 'dist'])

fig.suptitle("Testando", fontsize = 16, fontweight = 'bold');
```
